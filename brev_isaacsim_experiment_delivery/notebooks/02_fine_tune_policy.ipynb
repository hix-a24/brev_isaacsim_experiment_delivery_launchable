# 02 – Fine‑Tune Policy

This notebook demonstrates how to fine‑tune the vision–language–action (VLA) policy on the collected demonstrations.  In practice, you would integrate with the OpenVLA and Octo training pipelines.  Here we simulate training by iterating over the demonstrations and computing trivial loss values.

```python
from pathlib import Path
import numpy as np

# Load demonstrations (saved in the previous notebook)
demo_dir = Path('runs/demonstrations')
drawer_demos = np.load(demo_dir / 'drawer_demos.npy', allow_pickle=True)
clutter_demos = np.load(demo_dir / 'clutter_demos.npy', allow_pickle=True)

# Placeholder: simulate fine‑tuning of OpenVLA and Octo models
def fake_fine_tune(demos, model_name):
    total_loss = 0.0
    for traj in demos:
        # Compute a dummy loss as the sum of action magnitudes
+        loss = sum(np.linalg.norm(step['action']) for step in traj)
        total_loss += loss
    avg_loss = total_loss / len(demos)
    print(f'{model_name}: trained on {len(demos)} demos; average dummy loss={avg_loss:.3f}')
    return avg_loss

# Fine‑tune policies (dummy)
openvla_loss_drawer = fake_fine_tune(drawer_demos, 'OpenVLA (drawer)')
openvla_loss_clutter = fake_fine_tune(clutter_demos, 'OpenVLA (clutter)')
octo_loss_drawer = fake_fine_tune(drawer_demos, 'Octo (drawer)')
octo_loss_clutter = fake_fine_tune(clutter_demos, 'Octo (clutter)')

# Save dummy checkpoints (for demonstration)
checkpoints_dir = Path('checkpoints')
checkpoints_dir.mkdir(exist_ok=True)
(checkpoints_dir / 'openvla_pilot.ckpt').write_text('dummy checkpoint data')
(checkpoints_dir / 'octo_pilot.ckpt').write_text('dummy checkpoint data')
print('Saved dummy model checkpoints.')
```
