# 01 – Generate Demonstrations

This notebook illustrates how to collect demonstration trajectories for the tasks defined in the experiment.  Demonstrations consist of short tele‑operated or scripted interactions with the environment that show the policy how to achieve the tasks.  For the purposes of this repository, we generate dummy demonstrations using the placeholder task classes; in your deployment you would replace these calls with your tele‑operation interface or scripted policies.

```python
from pathlib import Path
from src.utils.config import load_config
from src.tasks import DrawerObjectTask, ClutterSortTask
import numpy as np

# Load pilot configuration (adjust path as needed)
config = load_config('configs/pilot_config.yaml')

# Create directories to store demonstrations
demo_dir = Path('runs/demonstrations')
demo_dir.mkdir(parents=True, exist_ok=True)

def collect_demos(task_name, task_class, num_demos):
    demos = []
    for i in range(num_demos):
        env = task_class(config=config['tasks'][task_name])
        obs = env.reset(shift_config={})
        done = False
        steps = 0
        trajectory = []
        while not done and steps < 50:
            # Generate a random action for demonstration purposes
            action = np.random.uniform(-0.1, 0.1, size=(7,), dtype=np.float32)
            next_obs, reward, done, info = env.step(action)
            trajectory.append({'obs': obs, 'action': action})
            obs = next_obs
            steps += 1
        env.close()
        demos.append(trajectory)
    return demos

# Collect demonstrations for each task (dummy)
drawer_demos = collect_demos('drawer', DrawerObjectTask, config['tasks']['drawer']['demo_count'])
clutter_demos = collect_demos('clutter_sort', ClutterSortTask, config['tasks']['clutter_sort']['demo_count'])

# Save demonstrations as numpy files (placeholder)
np.save(demo_dir / 'drawer_demos.npy', drawer_demos, allow_pickle=True)
np.save(demo_dir / 'clutter_demos.npy', clutter_demos, allow_pickle=True)

print(f'Collected {len(drawer_demos)} drawer demos and {len(clutter_demos)} clutter demos.')
```
